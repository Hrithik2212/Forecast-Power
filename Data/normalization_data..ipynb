```python
import pandas as pd
df = pd.read_csv("data_Florida_2005_2015._10.csv")
df
```

```python
from sklearn import preprocessing
mms = preprocessing.MinMaxScaler()
minmax = mms.fit_transform(df)
print(minmax)
labels =['avg_demand','Number_of_energy_intensive_industry','Number_of_energy_non_intensive_industry','gdp_growth','load_share','average_retail_sales']
df1 = pd.DataFrame.from_records(minmax, columns=labels)
df1.to_csv("norm_data_Florida_2005_2015_10.csv", sep="\t")
```
